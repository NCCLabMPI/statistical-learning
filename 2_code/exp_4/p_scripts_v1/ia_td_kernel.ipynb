```python
import pandas as pd
import numpy as np 
from sklearn.neighbors.kde import KernelDensity
from matplotlib.pyplot import plot
from scipy.signal import argrelextrema
```

```python
exp_3_rt = pd.read_csv("exp3array.csv")['rt']
rt_array = array(exp_3_rt).reshape(-1,1)
```

Frankly, I'm not quite sure what reshape is doing here... Return.

Anyway, now we fit the array of RTs to a gaussian kernel with bandwidth 3. 

```python
kde = KernelDensity(kernel='linear', bandwidth=3).fit(rt_array)
s = linspace(0,round(max(exp_3_rt))+1)
e = kde.score_samples(s.reshape(-1,1))
plot(s,e)
```

Hmm... so do we need a different kernel distribution? E.g. gaussian? 

```python
mi, ma = argrelextrema(e, np.less)[0], argrelextrema(e, np.greater)[0]
print("Minima:", s[mi])
print("Maxima:", s[ma])
```

```python
print(rt_array[rt_array < mi[0]], rt_array[(rt_array >= mi[0]) * (rt_array <= mi[1])], rt_array[rt_array >= mi[1]])
```

```python
plot(s[:mi[0]+1], e[:mi[0]+1], 'r',
     s[mi[0]:mi[1]+1], e[mi[0]:mi[1]+1], 'g',
     s[mi[1]:], e[mi[1]:], 'b',
     s[ma], e[ma], 'go',
     s[mi], e[mi], 'ro')
```
